### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.

```python
#assign to new variable
df_HangSeng = pd.read_csv("https://raw.githubusercontent.com/ridhodori/visualisasi-data-/main/Hang_Seng.csv") 
df_Nasdaq = pd.read_csv("https://raw.githubusercontent.com/ridhodori/visualisasi-data-/main/Nasdaq.csv") 
df_Nikkei = pd.read_csv("https://raw.githubusercontent.com/ridhodori/visualisasi-data-/main/Nikkei.csv") 
```

```python
df_HangSeng.head()
```

```python
df_Nasdaq.head()
```

```python
df_Nikkei.head()
```

#### Code

```python
# type your code here

#round up 2 number behind comma
df_HangSeng = df_HangSeng.round(2)
df_Nasdaq = df_Nasdaq.round(2)
df_Nikkei = df_Nikkei.round(2)
#drop the null
df_HangSeng.dropna(inplace = True, axis = 0)
df_Nasdaq.dropna(inplace = True, axis = 0)
df_Nikkei.dropna(inplace = True, axis = 0)
#change the datatype date 
df_HangSeng['Date'] = pd.to_datetime(df_HangSeng['Date'])
df_Nasdaq['Date'] = pd.to_datetime(df_HangSeng['Date'])
df_Nikkei['Date'] = pd.to_datetime(df_HangSeng['Date'])
df_HangSeng.head()
```

```python
df_Nasdaq.head()
```

```python
df_Nikkei.head()
```

```python
#sort the date only 2019
df_HangSeng.index = df_HangSeng["Date"]
df_HangSeng = df_HangSeng.loc['2019-01-01':'2019-12-31']
df_Nasdaq.index = df_Nasdaq["Date"]
df_Nasdaq = df_Nasdaq.loc['2019-01-01':'2019-12-31']
df_Nikkei.index = df_Nikkei["Date"]
df_Nikkei = df_Nikkei.loc['2019-01-01':'2019-12-31']
```

```python
#count day percent change with adj close and make new collumn to assign name day_per_change
df_HangSeng["Day_Perc_Change"] = df_HangSeng["Adj Close"].pct_change()*100
df_Nasdaq["Day_Perc_Change"] = df_Nasdaq["Adj Close"].pct_change()*100
df_Nikkei["Day_Perc_Change"] = df_Nikkei["Adj Close"].pct_change()*100
df_HangSeng.dropna(axis = 0, inplace = True)
df_Nasdaq.dropna(axis = 0, inplace = True)
df_Nikkei.dropna(axis = 0, inplace = True)
df_HangSeng.head()
```

```python
df_Nasdaq.head()
```

```python
df_Nikkei.head()
```

```python
#make the if else like the table 
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

```python
#search the trend from day_percent_change then assign to  new collumn name trend 
df_HangSeng["Trend"] = df_HangSeng["Day_Perc_Change"].apply(lambda x:trend(x))
df_Nasdaq["Trend"] = df_Nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))
df_Nikkei["Trend"] = df_Nikkei["Day_Perc_Change"].apply(lambda x:trend(x))
df_HangSeng.head()
```

```python
df_Nasdaq.head()
```

```python
df_Nikkei.head()
```

```python
#visualise the tree dataset with pie chart based on the trend result 
plt.subplot(1,3,1)
HangSeng_pie_data = df_HangSeng.groupby('Trend')
pie_label = sorted([i for i in df_HangSeng.loc[:, 'Trend'].unique()])
plt.pie(HangSeng_pie_data['Trend'].count(), labels = pie_label, autopct = '%1.1f%%', radius = 2)
plt.title("Hang Seng", y=1.5)

plt.subplot(1,3,2)
Nasdaq_pie_data = df_Nasdaq.groupby('Trend')
pie_label2 = sorted([i for i in df_Nasdaq.loc[:, 'Trend'].unique()])
plt.pie(Nasdaq_pie_data['Trend'].count(), labels = pie_label2, autopct = '%1.1f%%', radius = 2)
plt.title("Nasdaq", y=1.5)

plt.subplot(1,3,3)
Nikkei_pie_data = df_Nikkei.groupby('Trend')
pie_label3 = sorted([i for i in df_Nikkei.loc[:, 'Trend'].unique()])
plt.pie(Nikkei_pie_data['Trend'].count(), labels = pie_label3, autopct = '%1.1f%%', radius = 2)
plt.title("Nikkei", y=1.5)

plt.subplots_adjust(right=4.1, top=1)
plt.show()
```

#### Penjelasan 
In the three pie charts above, the Nasdaq stock had a very positive trend in 2019 and even got to the top even though it was only 0.4%, but the Nasdaq stock was also the worst/lowest with 1.6%. Meanwhile, Hang Seng shares in 2019 had a negative trend which was marked by a value of 11.8% where the negative value was the largest of the three stocks, but this stock was also not too negative with a slightly positive value of 14.7%. . Then the Nikkei stock in 2019 did not experience much change, this can be seen with an unchanged value of 52.9% which has the largest value of the three existing stocks, and this stock has also experienced a slight negative which is marked by a value of 14.6 %, this value is the second largest value on the Nikkei stock pie chart.

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
#make merge dataset 
df_merge = pd.concat([df_HangSeng["Adj Close"], df_Nasdaq["Adj Close"], df_Nikkei["Adj Close"]], axis=1)
df_merge.columns = ["Hang.Seng", "Nasdaq", "Nikkei"]
df_merge.dropna(inplace = True, axis = 0)
df_merge.head()
```

```python
#make the dataset change to persen and assign to new dataset 
df_perct = df_merge.pct_change()*100
df_perct.dropna(inplace = True, how = "any", axis = 0)
df_perct.head()
```

```python
#visualise the merge dataset 
volatil = df_perct[['Hang.Seng', 'Nasdaq', 'Nikkei']].rolling(7).std()*np.sqrt(7)
volatil.plot(figsize = (20, 10))
plt.title("Volatility of 3 stock prices in 2019")

plt.show()
```

#### Penjelasan 
as you can be seen from the visualization above that the Nasdaq stock is the most volatile stock so that the Nasdaq stock makes a stock that has a high profit but has the same amount of risk. while the Nikkei stock had a high volatility value in the early months of 2019 but by the end of the year the stock experienced a significant decrease in volatility. On the Hang Seng stock, the volatility value is stable, neither too high nor too low.
